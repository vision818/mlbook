```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression

```

```python
#Download and prepare the data

# data_root = "https://github.com/ageron/data/blob/main/lifesat/lifesat_full.csv"
# pd.read_csv(data_root)

data_url = "https://raw.githubusercontent.com/ageron/data/main/lifesat/lifesat_full.csv"

# Read the CSV file into a pandas DataFrame
lifesat = pd.read_csv(data_url)

# Display the first few rows of the DataFrame to verify that the data was loaded correctly
#print(data.head())
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

```

```python
#Visualize the data
lifesat.plot(kind = "scatter", grid = True, x = "GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 5, 8])
plt.show()
```

```python
#Select a linear model
model = LinearRegression()
```

```python
#Train the model
model.fit(X, y)
```

```python
#Make prediction for cyprus
X_new = [[37_655.2]]
print(model.predict(X_new))
```
